```python
import requests
import tkinter as tk
from tkinter import messagebox
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
API_KEY = "51b67468e45af15fefdbcf7f760a6370"
#current weather for a given city
def get_weather(city_name):
    BASE_URL="https://api.openweathermap.org/data/2.5/weather"
    url=f"{BASE_URL}?q={city_name}&appid={API_KEY}&units=metric"
    response=requests.get(url)
    if response.status_code==200:
        data=response.json()
        main=data['main']
        weather=data['weather'][0]
        return {
            "temperature": main['temp'],
            "humidity": main['humidity'],
            "pressure": main['pressure'],
            "description": weather['description']
        }
    else:
        return None

#10 coldest cities in Pakistan
def get_coldest_cities():
    cities=["Islamabad", "Karachi", "Lahore", "Peshawar", "Quetta", "Multan", "Faisalabad", "Rawalpindi", "Sialkot", "Hyderabad"]
    city_temps=[]
    for city in cities:
        weather=get_weather(city)
        if weather:
            city_temps.append((city, weather['temperature']))
    #sorting cities by temperature
    city_temps.sort(key=lambda x:x[1])
    return city_temps
#display current weather
def display_weather():
    city_name=city_entry.get()
    if not city_name:
        messagebox.showerror("input error","please enter a city name")
        return
    weather=get_weather(city_name)
    if weather:
        result_label.config(
            text=
        (
                f"City: {city_name.capitalize()}\n"
                f"Temperature: {weather['temperature']}°C\n"
                f"Humidity: {weather['humidity']}%\n"
                f"Pressure: {weather['pressure']} hPa\n"
                f"Weather: {weather['description'].capitalize()}"
            )
        )
    else:
        messagebox.showerror("eror","City not found or invalid API key")
#bar chart for 10 coldest cities
def display_coldest_cities():
    coldest_cities=get_coldest_cities()
    if not coldest_cities:
        messagebox.showerror("error","unable to fetch data for coldest cities")
        return
    #data for visualization
    cities=[city[0] for city in coldest_cities]
    temperatures=[city[1] for city in coldest_cities]
    #matplotlib figure
    fig=Figure(figsize=(6, 4), dpi=100)
    ax=fig.add_subplot(111)
    ax.bar(cities, temperatures, color='blue')
    ax.set_title("10 Coldest Cities in Pakistan")
    ax.set_ylabel("Temperature(°C)")
    ax.set_xlabel("Cities")
    #chart in the Tkinter GUI
    chart=FigureCanvasTkAgg(fig, master=root)
    chart.get_tk_widget().pack(pady=20)
    chart.draw()
#set up the GUI
root=tk.Tk()
root.title("Weather App")
root.geometry("500x600")
# city input
city_label=tk.Label(root, text="Enter City Name:", font=("Arial", 14))
city_label.pack(pady=10)
city_entry=tk.Entry(root, width=30, font=("Arial", 14))
city_entry.pack(pady=5)
#buttons
weather_button=tk.Button(root, text="Get Weather", command=display_weather, font=("Arial", 14))
weather_button.pack(pady=10)
coldest_cities_button=tk.Button(root, text="Show 10 Coldest Cities", command=display_coldest_cities, font=("Arial", 14))
coldest_cities_button.pack(pady=10)
#label to display weather results
result_label = tk.Label(root, text="", font=("Arial", 14), justify="left")
result_label.pack(pady=20)
#run GUI
root.mainloop()
```
