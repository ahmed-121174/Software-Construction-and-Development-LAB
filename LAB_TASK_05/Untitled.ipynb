```python
import requests
```

```python
response=requests.get("http://pwr.nu.edu.pk/cs-faculty/")
```

```python
from bs4 import BeautifulSoup as bs
```

```python
soup=bs(response.content)
```

```python
print(soup.prettify())
```

```python
soup.find_all("div", class_="faculty-member")
```

```python
faculty_h2=soup.select("div.faculty-member h2")
```

```python
for i in faculty_h2:
    print(i.text)
```

```python
p_faculty=soup.select("div.faculty-member h2, div.faculty-member p")
```

```python
for i in p_faculty:
    print(i.text)
```
